2. Data Preprocessing : contraction expansion, emoji to description text, removing punctuation, lemmatization will be preformed on here

````python
import string
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer
import contractions
import emoji

def wordTokenize(sentence):
  return word_tokenize(sentence)

def emoji2description(text):
  return emoji.replace_emoji(text, replace=lambda chars, data_dict: ' '.join(data_dict['en'].split('_')).strip(':')+" ")

def removePunctuation(words):
  result = [word for word in words if word not in string.punctuation]
  return result

def expandContraction(sentence):
  result = [contractions.fix(word) for word in sentence.split(' ')]
  return ' '.join(result)

def filterStopword(words):
  stop_words = stopwords.words('english')
  filtered_words = [word for word in words if word not in stop_words]
  return filtered_words

def stemming(words):
  porter = PorterStemmer()
  stemmed = [porter.stem(word) for word in words]
  return stemmed

def lemmatization(words): #to verb only
  lemmatizer= WordNetLemmatizer()
  lemmatized = [lemmatizer.lemmatize(word,pos='v') for word in words ]
  return lemmatized
def partOfSpeech(words):
  pos = pos_tag(words)
  return pos

def partOfSpeech_nouns(words):
  pos = pos_tag(words)
  nouns = [word for word, pos in pos if pos.startswith('N')]
  return nouns

def toString(words):
  return ' '.join(words)
````

````python
def step(sentence):
  sentence = expandContraction(sentence)  ##before remove punctuation
  sentence = emoji2description(sentence)  ##before convert lower case
  result = wordTokenize(sentence.lower())
  result = removePunctuation(result)
  result = lemmatization(result)
  result = toString(result)
  return result
````

````python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from scipy.special import softmax
# load model and tokenizer
roberta = "cardiffnlp/twitter-roberta-base-sentiment"

model = AutoModelForSequenceClassification.from_pretrained(roberta)
tokenizer = AutoTokenizer.from_pretrained(roberta)

labels = ['Negative', 'Neutral', 'Positive']
````

````python
def roberta_sentiment(text):
  try:
    encoded_tweet = tokenizer(text, return_tensors='pt')
    output = model(**encoded_tweet)

    scores = output[0][0].detach().numpy()
    scores = softmax(scores)
    return scores[0],scores[1],scores[2]
  except Exception as e:
    print(f"Error occurred while processing {text}: {str(e)}")
    return 0,0,0
````

````python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch.nn.functional as F
from transformers import pipeline

# Load Aspect-Based Sentiment Analysis model
absa_tokenizer = AutoTokenizer.from_pretrained("yangheng/deberta-v3-base-absa-v1.1")
absa_model = AutoModelForSequenceClassification.from_pretrained("yangheng/deberta-v3-base-absa-v1.1")
````

````python
import numpy as np
def absa_sentiment(sentence):
  nouns = partOfSpeech_nouns(wordTokenize(sentence))
  neg = []
  neu = []
  pos = []
  for i in nouns:
    inputs = absa_tokenizer(f"[CLS] {sentence} [SEP] {i} [SEP]", return_tensors="pt")
    outputs = absa_model(**inputs)
    probs = F.softmax(outputs.logits, dim=1)
    probs = probs.detach().numpy()[0]
    index = np.argmax(probs)
    match index:
      case 0:
        neg.append((i,max(probs)))
      case 1:
        neu.append((i,max(probs)))
      case 2:
        pos.append((i,max(probs)))
  return neg,neu,pos
````
